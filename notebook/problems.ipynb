## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import scipy.stats as stats

# Define the functions


def normal_dist (x, media = 0, desviacin = 1):

    # Probability density function

    pdf_fnc = stats.norm.pdf(x, loc = media, scale = desviacin)

    return pdf_fnc



def normal_dist_c (x, media = 0, desviacin = 1):

    # Cumulative density function

    cdf_fnc = stats.norm.cdf(x, loc = media, scale = desviacin)

    return cdf_fnc


def normal_dist_p (x, media = 0, desviacin = 1):

    # Percent pont function

    ppf_fnc = stats.norm.ppf(x, loc = media, scale = desviacin)

    return ppf_fnc


def normal_dist_r (n, media = 0, desviacin = 1):
    
    # Random variantes
     
    rvs_fnc = stats.norm.rvs(size = n, loc = media, scale = desviacin)
    
    return rvs_fnc




```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(24)

# Define values for normal distribution
media = 0
desviacin = 1

# Define x-axe
data = np.linspace(-5, 5, 1000)

# Call the functions into variables
cdf = normal_dist_c(data, media = media, desviacin = desviacin)
ppf = normal_dist_p(data, media = media, desviacin = desviacin)
rvs = normal_dist_r(1000, desviacin = desviacin)

# Display the 3 in one plot
fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, cdf, "r-")
axis[0].set_title('CDF')

axis[1].plot(data, ppf, "b-")
axis[1].set_title('PPF')

axis[2].plot(data, rvs, "g-")
axis[2].set_title('RVS')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Definition of the functions


def uniform_dist (x, low = 0, high = 1):

    # Probability density function

    pdf_fnc_u = stats.uniform.pdf(x, low = 0, scale = (high - low))
    
    return pdf_fnc_u


def uniform_dist_c (x, low = 0, high = 1):

    # Cumulative density function
    
    cdf_fnc_u = stats.uniform.cdf(x, loc = low, scale = (high - low))
    
    return cdf_fnc_u


def uniform_dist_p (x, low = 0, high = 1):

    # Percent point function

    ppf_fnc_u = stats.uniform.ppf(x, loc = low, scale = (high - low))
    
    return ppf_fnc_u


def uniform_dist_r (x, low = 0, high = 1):

    # Random variates

    ppf_rvs_u = stats.uniform.rvs(size = x, loc = low, scale = (high - low))
    
    return ppf_rvs_u
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(24)

# Set parameters for uniform distribution
low = 0
high = 1

# Set the x-axe
data = np.linspace(-5, 5, 1000)

# Import the funtions into variables
cdf = uniform_dist_c(data, low = low, high = high)
ppf = uniform_dist_p(data, low = low, high = high)
rvs = uniform_dist_r(1000, low = low, high = high)


# Display the functions into one plot
fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].plot(data, cdf, "r-")
axis[0].set_title('CDF')

axis[1].plot(data, ppf, "b-")
axis[1].set_title('PPF')

axis[2].plot(data, rvs, "g-")
axis[2].set_title('RVS')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
